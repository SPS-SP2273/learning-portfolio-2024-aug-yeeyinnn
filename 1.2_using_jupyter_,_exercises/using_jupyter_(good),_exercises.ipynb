#### what is each of the numbers in the %%timeit output represents
**Runs** means how many times the entire code ran <br>
**Loops** means how many times the code executed in each run<br>
**Mean** is the average time used for each loops which followed by the **standard deviation** ($\pm$ __ )<br>
Large number of loops is to get more accurate measurement<br>
